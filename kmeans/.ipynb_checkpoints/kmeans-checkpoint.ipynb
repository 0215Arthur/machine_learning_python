```python
from kmeans_base import KMeansBase
from kmeans_plus import KMeansPlusPlus
import time

from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn import datasets

import numpy as np

from utils.misc_utils import distance, sortLabel
from kmeans.kmeans_base import KMeansBase

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
allData = {"iris":datasets.load_iris(), "boston":datasets.load_boston(), "digits":datasets.load_digits()}
models = {"KmeansBase":KMeansBase(3), "kmeans++":KMeansPlusPlus(3, init="k-means++"), "sklearnKmeans":KMeans(init='k-means++', n_clusters= 3)}
```

```python
for dataName, value in allData.items():
    #降维
    tsne = TSNE(n_components=2, init='pca', random_state=0)
    X_tsne = tsne.fit_transform(value.data)
    for modelName, model in models.items():
        startTime = time.time()
        labels = model.fit_predict(value.data)
        takeTime = time.time() - startTime
        print("dataset=%s model=%s takeTime=%s"%(dataName, modelName, round(takeTime, 5)))
        plt.scatter(X_tsne[:, 0],X_tsne[:, 1], c=labels)
        plt.show()
    print("-" * 100)
    
    
```
